```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
csv_path = "Resources/avg_rain_state.csv"

avg_state_rain = pd.read_csv(csv_path)
avg_state_rain
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(avg_state_rain))
tick_location = [x+0.5 for x in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, avg_state_rain["Inches"], color='b', alpha=0.5)
plt.xticks(tick_location, avg_state_rain["State"], rotation="vertical")
```

```python
# Set x and y limits
plt.xlim(0, len(x_axis))
plt.ylim(0, max(avg_state_rain["Inches"])+20)
```

```python
# Set a Title and labels
plt.title("Average Rain per State")
plt.xlabel("State")
plt.ylabel("Rain (Inches)")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("Images/avg_state_rain.png")
plt.show()
```
